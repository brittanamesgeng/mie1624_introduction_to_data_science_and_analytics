```python

def page_columbia():
        """Analyse single query result link

        Analyse link from self.__link. First, find every job information by
        tag class name or tag id. And then analyse tag block by div_analyse.

        Args:
            link: link from self.__link
        
        """
        link = 'http://datascience.columbia.edu/course-inventory'
        print('Analysing link: ' + link + '\nPlease wait...')
        content = urlopen(link).read()
        soup = BeautifulSoup(content, 'lxml')
        course_div = soup.find_all('div',{'class':'course'})
        course_list=[]
        for course in course_div:
            title = course.h3.get_text().strip(' \t\n\r')
            summary = course.select("p:nth-of-type(3)")[0].get_text().strip( '\t\n\r')
            college = 'columbia university'
            country ='USA'

            course_list.append([title,summary,college,country])
        print('Finish '+ link)
        return course_list
```
